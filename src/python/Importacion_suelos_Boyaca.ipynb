```python
import re
import pandas as pd
from datetime import datetime 

fecha = datetime.today().strftime('%Y%m%d')
num_format = re.compile("^[\-]?[0-9]*\.?[0-9]+$")
```

```python
df = pd.read_csv("D:\AEPS 2.0 Boyaca\datos\Suelos\data\Suelos_600muestras_N_CO_v2.csv", encoding='ISO-8859-1', thousands=',', usecols=lambda c: not c.startswith('Unnamed:'))

```

```python
#soc_people
#'kind_document'	'document'	'name'	'last_name'	'cellphone'	'sex'	'municipality'	'address'	'email'	'ext_id'

documents = df['GUIA']
last_names = df['ID_PUNTOS']
municipalityes = df['MUNICIPIO'].str.strip() #Se eliminan los espacios al inicio y final para evitar errores de realcion
address = df['MUNICIPIO']
codigos = df['ID']

d = {'kind_document':'O','document': documents, 'name': 'Muestra Suelo','last_name':last_names,
'cellphone': 1,'sex': 'O', 'municipality': municipalityes,'address': address, 'email': '','ext_id': codigos}

soc_people = pd.DataFrame(d)

soc_people.to_csv('soc_people.csv',index=False,  encoding='utf-8-sig')
```

```python
#far_farms
#'name'	'latitude'	'location_comments'	'ext_id'	'farmer'	'longitude'

latitude_fram = df['Lati']
longitude_fram = df['Long']
finca = 'Finca de ' + df['ID']
ext_id_fram = 'FINCA-' + df['ID'] #Se genera codigo de la finca con el codigo del agricultor

f = {'name': finca,'latitude': latitude_fram, 'location_comments': '','ext_id' : ext_id_fram,'farmer': codigos,'longitude': longitude_fram}

far_farms = pd.DataFrame(f)
far_farms = far_farms.drop_duplicates() #Se eliminan los duplicados

far_farms.to_csv('far_farms.csv', index=False, encoding='utf-8-sig')
```

```python
#far_plots
#'farm'	'name'	'latitude'	'altitude'	'ext_id'	'longitude'

ext_id_plot = 'LOTE-' + df['ID'] #Se genera codigo del lote con el codigo del agricultor
lote = 'Lote ' +  df['ID']

p = {'farm': ext_id_fram, 'name': lote, 'latitude': latitude_fram, 'altitude': 0, 'ext_id': ext_id_plot, 'longitude': longitude_fram}

far_plots = pd.DataFrame(p)
far_plots = far_plots.drop_duplicates() #Se eliminan los duplicados

far_plots.to_csv('far_plots.csv', index=False, encoding='utf-8-sig')
```

```python
#far_production_events
# 'technical'	'ext_id'	'plot'	'form'


df['event'] = 'EVENTO-SUE-' + fecha + '-' +df['ID'] #Se genera los id externos para los eventos productivos

e = {'technical' : 1, 'ext_id' : df['event'], 'plot': ext_id_plot, 'form' : 1}

event = pd.DataFrame(e)

event.to_csv('far_production_events.csv', index=False, encoding='utf-8-sig')
```
